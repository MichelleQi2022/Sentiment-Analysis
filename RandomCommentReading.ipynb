```python
import csv
import os
import xlwt
import random
from snownlp import SnowNLP

global totnum
totnum = 0
global cnt
cnt = 0
random_list=[]

def eachfile(filepath):
    files=os.listdir(filepath)
    for fi in files:
        with open('D:/抑郁症医生评论/'+fi,'r',encoding='utf-8') as csvfile:
            reader = csv.DictReader(csvfile)
            comment_list = [row['患者评论'] for row in reader]
        for i in comment_list:
            global totnum
            totnum+=1
            if totnum in random_list:
                print(i)
                global cnt
                cnt+=1
                worksheet.write(cnt,0,cnt,style)
                worksheet.write(cnt,1,i,style)
                workbook.save('随机读取1.xls')
                s=SnowNLP(i)
                print(s.sentiments)
                if s.sentiments>=0.75:
                    worksheet.write(cnt,2,'Pos',style)
                    workbook.save('随机读取1.xls')
                elif s.sentiments<0.5:
                    worksheet.write(cnt,2,'Neg',style)
                    workbook.save('随机读取1.xls')
                else:
                    worksheet.write(cnt,2,'Neu',style)
                    workbook.save('随机读取1.xls')

random_list=random.sample(range(1,11021+1),25)
workbook = xlwt.Workbook(encoding='utf-8')
worksheet = workbook.add_sheet('抑郁症医生评论')
style = xlwt.XFStyle()
font = xlwt.Font()
font.name = '等线'
font.height = 220
font.bold = False
font.underline = False
font.italic = False
style.font = font
worksheet.write(0,0,'序号',style)
worksheet.write(0,1,'评论',style)
worksheet.write(0,2,'程序结果',style)
worksheet.write(0,3,'人工结果',style)
workbook.save('随机读取1.xls')
eachfile("D:\\抑郁症医生评论")
```

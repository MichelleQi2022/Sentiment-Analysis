```python
import os
import csv
import xlwt
import jieba
import jieba.analyse
from snownlp import SnowNLP

global totnum
totnum = 0

def split(st):
    ids,names = "",""
    for i in st:
        if i.isdigit():
            ids = ids+i
        elif i=='.':
            break
        else:
            names = names+i
    worksheet.write(totnum,0,ids,style)
    worksheet.write(totnum,1,names,style)
    workbook.save('评论数据v3.xls')

def eachfile(filepath):
    files=os.listdir(filepath)
    for fi in files:
        global totnum
        totnum+=1
        split(fi)
        with open('D:/抑郁症医生评论/'+fi,'r',encoding='utf-8') as csvfile:
            reader = csv.DictReader(csvfile)
            comment_list = [row['患者评论'] for row in reader]
            
            
        #list=f.readlines()
        #sentimentslist = []
        
        text_string = "".join(t for t in comment_list)
        keywords = jieba.analyse.extract_tags(text_string, 8)
        
        #print("【关键词】", keywords)
        
        seg_list = jieba.lcut(text_string, cut_all=False)
        word_freq = {}
        for word in seg_list:
            if word in word_freq:
                word_freq[word] += 1
            else:
                word_freq[word] = 1

        freq_word = []
        for word, freq in word_freq.items():
            if len(word) != 1:   
                freq_word.append((word, freq))
        freq_word.sort(key=lambda x: x[1], reverse=True)
        cnt = 0
        for word, freq in freq_word[:30]:
            cnt+=1
            if cnt==1:
                worksheet.write(totnum,7,word,style)
                worksheet.write(totnum,8,freq,style)
                workbook.save('评论数据v3.xls')
            elif cnt==2:
                worksheet.write(totnum,9,word,style)
                worksheet.write(totnum,10,freq,style)
                workbook.save('评论数据v3.xls')
            elif cnt==3:
                worksheet.write(totnum,11,word,style)
                worksheet.write(totnum,12,freq,style)
                workbook.save('评论数据v3.xls')
            elif cnt==4:
                worksheet.write(totnum,13,word,style)
                worksheet.write(totnum,14,freq,style)
                workbook.save('评论数据v3.xls')
            elif cnt==5:
                worksheet.write(totnum,15,word,style)
                worksheet.write(totnum,16,freq,style)
                workbook.save('评论数据v3.xls')
                break

        count=0
        sentiments_tot=0
        sentiments_ave=0
        pos=0
        neg=0
        neu=0
        
        #f.close()
        
        for i in comment_list:
            count+=1
            s=SnowNLP(i)
            sentiments_tot += s.sentiments
            if s.sentiments>=0.75:
                pos+=1
            elif s.sentiments<0.5:
                neg+=1
            else:
                neu+=1
                
            #sentimentslist.append(s.sentiments)
            
        if count!=0:
            sentiments_ave = sentiments_tot / count
        worksheet.write(totnum,2,sentiments_ave,style)
        worksheet.write(totnum,3,count,style)
        worksheet.write(totnum,4,pos,style)
        worksheet.write(totnum,5,neg,style)
        worksheet.write(totnum,6,neu,style)
        workbook.save('评论数据v3.xls')
def main():
    eachfile("D:\\抑郁症医生评论")
        
workbook = xlwt.Workbook(encoding='utf-8')
worksheet = workbook.add_sheet('抑郁症医生评论')
style = xlwt.XFStyle()
font = xlwt.Font()
font.name = '等线'
font.height = 220
font.bold = False
font.underline = False
font.italic = False
style.font = font
worksheet.write(0,0,'ID',style)
worksheet.write(0,1,'姓名',style)
worksheet.write(0,2,'Sentiments均值',style)
worksheet.write(0,3,'评论总数',style)
worksheet.write(0,4,'好评条数',style)
worksheet.write(0,5,'差评条数',style)
worksheet.write(0,6,'中评条数',style)
worksheet.write(0,7,'第一高频词',style)
worksheet.write(0,8,'频数',style)
worksheet.write(0,9,'第二高频词',style)
worksheet.write(0,10,'频数',style)
worksheet.write(0,11,'第三高频词',style)
worksheet.write(0,12,'频数',style)
worksheet.write(0,13,'第四高频词',style)
worksheet.write(0,14,'频数',style)
worksheet.write(0,15,'第五高频词',style)
worksheet.write(0,16,'频数',style)
workbook.save('评论数据v3.xls')
main()
```
